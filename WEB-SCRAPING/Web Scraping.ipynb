```python
pg2=requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")
```

```python
pg2
```

```python
soup2=BeautifulSoup(pg2.content,'html.parser')
```

```python
movie_tag=soup2.find('td',  class_='titleColumn')
movie_title=movie_tag.find('a')
```

```python
movie_title=[]
movie_year=[]
movie_rating=[]
num=0
for i in soup2.find_all('td',  class_='titleColumn'):
    title=i.find('a')
    movie_title.append(title.text)
    year=i.find('span', class_='secondaryInfo')
    movie_year.append(year.text)
    num=num+1
    if num==100:
        break
num=0
for i in soup2.find_all('td', class_='ratingColumn imdbRating'):
    rating=i.find('strong')
    movie_rating.append(rating.text)
    num=num+1
    if num==100:
        break
```

```python
print(len(movie_title),len(movie_year),len(movie_rating))
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Movie Title':movie_title,'Rating':movie_rating,'Year of release':movie_year})
df
```

b) Top 10 ODI Batsmen along with the records of their team and rating.
c) Top 10 ODI bowlers along with the records of their team and rating.

```python
pg5a=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
pg5a
```

```python
soup5a=BeautifulSoup(pg5a.content,'html.parser')
```

```python
cname=[]
matches=[]
points=[]
rating=[]

matches.append(soup5a.find('td', class_='rankings-block__banner--matches').text)
points.append(soup5a.find('td', class_='rankings-block__banner--points').text)
rating.append(soup5a.find('td', class_='rankings-block__banner--rating u-text-right').text.strip())

num=0
for i in soup5a.find_all('span',class_='u-hide-phablet'):
    cname.append(i.text)
    num=num+1
    if num==10:
        break
num=2
for i in soup5a.find_all('td', class_='table-body__cell u-center-text'):
    if(num%2==0):
        matches.append(i.text)
    else:
        points.append(i.text)
    num=num+1
    if num==20:
        break

num=1
for i in soup5a.find_all('td', class_='table-body__cell u-text-right rating'):
    rating.append(i.text)
    num=num+1
    if num==10:
        break
```

```python
df=pd.DataFrame({'Country Name':cname,'Matches':matches,'Points':points,'Rating':rating})
df
```

# 7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
pg7=requests.get('https://coreyms.com')
```

```python
pg7
```

```python
soup7=BeautifulSoup(pg7.content,'html.parser')
```

```python
heading=[]
date=[]
content=[]
video=[]

for i in soup7.find_all('a',class_='entry-title-link'):
    heading.append(i.text)
for i in soup7.find_all('time',class_='entry-time'):
    date.append(i.text)
for i in soup7.find_all('div',class_='entry-content'):
    cont=i.find('p')
    content.append(cont.text)
    try:
        vid=i.find('span',class_='embed-youtube')
        link=vid.find('iframe',class_='youtube-player')
        video.append(link['src'].split("?")[0].split("/")[-1])
    except:
        video.append('NA')
```

```python
len(heading),len(date),len(content),len(video)
```

```python
df=pd.DataFrame({'Heading':heading,'Date':date,'Content':content,'Video link':video})
df
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL 

```python
pg9=requests.get('https://www.dineout.co.in/mumbai-restaurants?search_str=mumbai')
```

```python
pg9
```

```python
soup9=BeautifulSoup(pg9.content)
```

```python
rname=[]
cuisine=[]
location=[]
rating=[]
image_url=[]

for i in soup9.find_all('a',class_='restnt-name ellipsis'):
    rname.append(i.text)
for i in soup9.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(i.text.split("|")[1])
for i in soup9.find_all('div',class_='restnt-loc ellipsis'):
    location.append(i.text)
for i in soup9.find_all('div',class_='img-wrap'):
    rating.append(i.text.strip()[-3:])
for i in soup9.find_all('img',class_='no-img'):
    image_url.append(i['data-src'])
```

```python
df=pd.DataFrame({'Restaurant Name':rname,'Cuisine':cuisine,'Location':location,'Rating':rating,'Image url':image_url})
df
```

# 10) Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/women-tshirts?ga_q=tshirts . 

```python
pg10=requests.get("https://www.bewakoof.com/women-tshirts?ga_q=tshirts")
```

```python
pg10
```

```python
soup10=BeautifulSoup(pg10.content)
```

```python
pname=[]
price=[]
img_url=[]

for i in soup10.find_all('div',class_='productCardDetail'):
    n=i.find('div')
    name=n.find('h3')
    pname.append(name.text)
    
for i in soup10.find_all('span',class_='discountedPriceText'):
    price.append(i.text)
    
for i in soup10.find_all('img',class_='productImgTag'):
    img_url.append(i['src'])
```

```python
#trying to store all list in one for loop
pname=[]
price=[]
img_url=[]

for i in soup10.find_all('div',class_='productCardBox'):
    pdeatils=i.find('div',class_='productCardDetail')
    n=pdeatils.find('div')
    name=n.find('h3')
    pname.append(name.text)
    p=pdeatils.find('div',class_='productPriceBox clearfix')
    pr=p.find('span',class_='discountedPriceText')
    price.append(pr.text)
    im=i.find('div',class_='productCardImg false')
    ima=im.find('div')
    imag=ima.find('img',class_='productImgTag')
    img_url.append(imag['src'])
```

```python
df=pd.DataFrame({'Product Name':pname,'Price':price,'Image url':img_url})
df
```

```python
df=pd.DataFrame({'Product Name':pname,'Price':price,'Image url':img_url})
df
```
