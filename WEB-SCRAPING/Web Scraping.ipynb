```python
pg4=requests.get("https://meesho.com/bags-ladies/pl/p7vbp")
```

```python
pg4
```

```python
soup4=BeautifulSoup(pg4.content,'html.parser')
```

```python
link='Card__BaseCard-sc-b3n78k-0 dUNFgg NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 fqvaeS'
pname_tag=soup4.find('div', class_=link)
pname=pname_tag.find('p')
pprice=pname_tag.find('h5')
```

```python
print(pname.text,pprice.text)
```

```python
link2='Card__BaseCard-sc-b3n78k-0 iJCKg NewProductCard__DiscountRow-sc-j0e7tu-15 kUcVjG NewProductCard__DiscountRow-sc-j0e7tu-15 kUcVjG'
pdiscount_tag=soup4.find('div',class_=link2)
pdiscount=pdiscount_tag.find('p')
pdiscount.text.split()[0]
```

```python
pname=[]
pprice=[]
pdiscount=[]
for i in soup4.find_all('div', class_=link):
    name=i.find('p')
    pname.append(name.text)
    price=i.find('h5')
    pprice.append(price.text)
    
for i in soup4.find_all('div', class_=link2):
    discount=i.find('p')
    pdiscount.append(discount.text.split()[0])
```

```python
df=pd.DataFrame({'Product Name':pname,'Product Price':pprice,'Product Discount':pdiscount})
df
```
