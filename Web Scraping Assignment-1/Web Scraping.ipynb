```python
pg2=requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")
```

```python
pg2
```

```python
soup2=BeautifulSoup(pg2.content,'html.parser')
```

```python
movie_tag=soup2.find('td',  class_='titleColumn')
movie_title=movie_tag.find('a')
```

```python
movie_title=[]
movie_year=[]
movie_rating=[]
num=0
for i in soup2.find_all('td',  class_='titleColumn'):
    title=i.find('a')
    movie_title.append(title.text)
    y=i.find('span', class_='secondaryInfo')
    a=y.text.replace('(','')
    movie_year.append(a.replace(')','')) 
    num=num+1
    if num==100:
        break
num=0
for i in soup2.find_all('td', class_='ratingColumn imdbRating'):
    rating=i.find('strong')
    movie_rating.append(rating.text)
    num=num+1
    if num==100:
        break
```

```python
print(len(movie_title),len(movie_year),len(movie_rating))
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Movie Title':movie_title,'Rating':movie_rating,'Year of release':movie_year})
df
```

# 4) Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm


```python
pg4=requests.get("https://presidentofindia.nic.in/former-presidents.htm")
```

```python
pg4
```

```python
soup4=BeautifulSoup(pg4.content,'html.parser')
```

```python
pname=[]
tof=[]

for i in soup4.find_all('div', class_='presidentListing'):
    name=i.find('h3')
    pname.append(name.text.split('(')[0])
    term=i.find('p')
    tof.append(term.text.split(':')[1])
```

# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
b) Top 10 ODI Batsmen along with the records of their team and rating.
c) Top 10 ODI bowlers along with the records of their team and rating.

```python
pg5a=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
pg5a
```

```python
soup5a=BeautifulSoup(pg5a.content,'html.parser')
```

```python
cname=[]
matches=[]
points=[]
rating=[]

matches.append(soup5a.find('td', class_='rankings-block__banner--matches').text)
points.append(soup5a.find('td', class_='rankings-block__banner--points').text)
rating.append(soup5a.find('td', class_='rankings-block__banner--rating u-text-right').text.strip())

num=0
for i in soup5a.find_all('span',class_='u-hide-phablet'):
    cname.append(i.text)
    num=num+1
    if num==10:
        break
num=2
for i in soup5a.find_all('td', class_='table-body__cell u-center-text'):
    if(num%2==0):
        matches.append(i.text)
    else:
        points.append(i.text)
    num=num+1
    if num==20:
        break

num=1
for i in soup5a.find_all('td', class_='table-body__cell u-text-right rating'):
    rating.append(i.text)
    num=num+1
    if num==10:
        break
```

```python
df=pd.DataFrame({'Country Name':cname,'Matches':matches,'Points':points,'Rating':rating})
df
```

# 7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :
i) Headline
ii) Time
iii) News Link

```python
pg7=requests.get('https://www.cnbc.com/world/?region=world')
```

```python
pg7 
```

```python
soup7=BeautifulSoup(pg7.content,'html.parser')
```

```python
headline=[]
time=[]
link=[]

for i in soup7.find_all('a',class_='LatestNews-headline'):
    headline.append(i.text)
    link.append(i['href'])
for i in soup7.find_all('time',class_='LatestNews-timestamp'):
    time.append(i.text)

for i in soup7.find_all('div',class_='RiverPlusCard-cardLeft'):
    h=i.find('div',class_='RiverHeadline-headline RiverHeadline-hasThumbnail')
    hh=h.find('a')
    headline.append(hh.text)
    link.append(hh['href'])
for i in soup7.find_all('div',class_='RiverPlusCard-cardLeft'):
    try:
        t=i.find('span',class_='RiverByline-datePublished')
        time.append(t.text)
    except:
        time.append('NA')
```

```python
len(headline)
```

```python
df=pd.DataFrame({'Headline':headline,'Time':time,'Link':link})
df
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL 

```python
pg9=requests.get('https://www.dineout.co.in/mumbai-restaurants?search_str=mumbai')
```

```python
pg9
```

```python
soup9=BeautifulSoup(pg9.content)
```

```python
rname=[]
cuisine=[]
location=[]
rating=[]
image_url=[]

for i in soup9.find_all('a',class_='restnt-name ellipsis'):
    rname.append(i.text)
for i in soup9.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(i.text.split("|")[1])
for i in soup9.find_all('div',class_='restnt-loc ellipsis'):
    location.append(i.text)
for i in soup9.find_all('div',class_='img-wrap'):
    rating.append(i.text.strip()[-3:])
for i in soup9.find_all('img',class_='no-img'):
    image_url.append(i['data-src'])
```

```python
df=pd.DataFrame({'Restaurant Name':rname,'Cuisine':cuisine,'Location':location,'Rating':rating,'Image url':image_url})
df
```

# 10) Write a python program to scrape the details of top publications from Google Scholar from https://scholar.google.com/citations?view_op=top_venues&hl=en
i) Rank
ii) Publication
iii) h5-index
 iv) h5-median

```python
pg10=requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
```

```python
pg10
```

```python
soup10=BeautifulSoup(pg10.content)
```

```python
rank=[]
publication=[]
h5index=[]
h5median=[]

for i in soup10.find_all('td',class_='gsc_mvt_p'):
    rank.append(i.text)
    
for i in soup10.find_all('td',class_='gsc_mvt_t'):
    publication.append(i.text)
    
for i in soup10.find_all('a',class_='gs_ibl gsc_mp_anchor'):
    h5index.append(i.text)
    
for i in soup10.find_all('span',class_='gs_ibl gsc_mp_anchor'):
    h5median.append(i.text)
  
```

```python
df=pd.DataFrame({'Rank':rank,'Publication':publication,'h5-index':h5index,'h5-median':h5median})
df
```
