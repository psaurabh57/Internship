# 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#extracting required urls in a particular list
urls=[]

start=0
end=3
for i in range(start,end):
    time.sleep(2)
    url=driver1.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']")
    for i in url:
        urls.append(i.get_attribute('href'))
    try:
        nb=driver1.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
        nb.click()
    except:
        break
```

```python
len(urls)
```

```python
#extracting required details in a particular list
brand_name=[]
product_name=[]
price=[]
return_exchange=[]
expected_delivery=[]
availability=[]

for i in urls:
    driver1.get(i)
    try:
        brand=driver1.find_element_by_xpath("//*[@id='productOverview_feature_div']/div/table/tbody/tr[1]/td[2]/span")
        brand_name.append(brand.text)
    except:
        brand_name.append("-")
    try:
        pname=driver1.find_element_by_id("productTitle")
        product_name.append(pname.text)
    except:
        product_name.append("-")
    try:
        try:
            pr=driver1.find_element_by_xpath("//span[@class='a-price a-text-price a-size-medium apexPriceToPay']")
            price.append(pr.text)
        except:
            pr=driver1.find_element_by_xpath("//span[@class='a-price aok-align-center priceToPay']")
            price.append(pr.text.replace('\n','.'))
    except:
        price.append("-")
    try:
        re=driver1.find_element_by_xpath("//*[@id='RETURNS_POLICY']/span/div[2]/a")
        return_exchange.append(re.text)
    except:
        return_exchange.append("-")
    try:
        ed=driver1.find_element_by_xpath('//*[@id="ddmDeliveryMessage"]/b')
        expected_delivery.append(ed.text)
    except:
        expected_delivery.append("-")
    try:
        try:
            radio_button=driver1.find_element_by_xpath("//*[@id='newAccordionRow']/div/div[1]/a/i")
            radio_button.click()
            time.sleep(2)
            ava=driver1.find_element_by_xpath("//*[@id='availability']/span")
            availability.append(ava.text)
        except:
            ava=driver1.find_element_by_xpath("//*[@id='availability']/span")
            availability.append(ava.text)
    except:
        availability.append("-")
```

```python
#creating dataframe and storing data in it
df1=pd.DataFrame({'Brand Name':brand_name,'Product Name':product_name,'Price':price,'Return/Exchange':return_exchange,'Expected Delivery':expected_delivery,'Availability':availability,'Product Url':urls})
df1.sample(10)
```

```python
#extracting data in csv formate
df1.to_csv('amazon_'+search+'_data.csv')
```

```python
#extracting required urls in a particular list 
urls=[]
hostel_name=[]
distance=[]
rating=[]
t_reviews=[]
o_review=[]
p_price=[]
d_price=[]
facilities=[]
description=[]


start=0
end=3
for i in range(start,end):
    time.sleep(2)
    
    for url in driver10.find_elements_by_xpath("//a[@class='view-button']"):
        urls.append(url.get_attribute('href'))
        
    try:
        for hname in driver10.find_elements_by_xpath('//h2[@class="title title-6"]'):
            hostel_name.append(hname.text)
    except:
        hostel_name.append('-')
        
    try:
        for dist in driver10.find_elements_by_xpath('//span[@class="description"]'):
            distance.append(dist.text.split('-')[1].split()[0])
    except:
        distance.append('-')
        
    try:
        for r in driver10.find_elements_by_xpath('//div[@class="rating rating-summary-container big"]'):
            rat=r.text.split('\n')
            if len(rat)==3:
                rating.append(rat[0])
            else:
                rating.append('-')            
    except:
        rating.append('-')
        
    try:
        for tr in driver10.find_elements_by_xpath('//div[@class="reviews"]'):
            t_reviews.append(tr.text)
    except:
        t_reviews.append('-')
        
    try:
        for o_r in driver10.find_elements_by_xpath('//div[@class="rating rating-summary-container big"]/div/div/span'):
            o_review.append(o_r.text)
    except:
        o_review.append('-')
        
    try:
        for p in driver10.find_elements_by_xpath('//div[@class="prices-col"]/a/div[1]'):
            pr=p.text.split('\n')
            if pr[0]=='Privates From':
                p_pr=pr[1].split()
                try:
                    p_price.append(p_pr[1])
                except:
                    p_price.append(p_pr[0])
            else:
                p_price.append('-')
    except:
        p_price.append('-')
        
    try:
        for p in driver10.find_elements_by_xpath('//div[@class="prices-col"]/a/div[2]'):
            pr=p.text.split('\n')
            if pr[0]=='Dorms From':
                p_pr=pr[1].split()
                try:
                    d_price.append(p_pr[1])
                except:
                    d_price.append(p_pr[0])
            else:
                d_price.append('-')
    except:
        d_price.append('-')
        
    try:
        next_btn=driver10.find_element_by_xpath('//div[@class="pagination-item pagination-next"]')
        next_btn.click()
    except:
        break
        
        
        
for url in urls:
    driver10.get(url)
    
    try:
        pd_btn=driver10.find_element_by_xpath('//*[@id="__layout"]/div/div[1]/section/div[6]/div/div[2]/div/div/a')
        pd_btn.click()  
        try:
            pd=driver10.find_element_by_xpath('//*[@id="__layout"]/div/div[1]/section/div[6]/div/div[2]/div/div/div[2]')
            description.append(pd.text)       
        except:
            description.append('-')                 
    except:
        try:
            pd=driver10.find_element_by_xpath('//*[@id="__layout"]/div/div[1]/section/div[6]/div/div[2]/div/div/div[2]')
            description.append(pd.text)       
        except:
            description.append('-')               
     
    try:
        fac=''
        for f in driver10.find_elements_by_xpath('//ul[@class="facility-sections"]/li/ul/li'):
            fac=fac+','+f.text
        facilities.append(fac)
    except:
        facilities.append('-')
            
```
